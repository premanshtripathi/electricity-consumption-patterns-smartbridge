```python
import pandas as pd

# 1. Load the data
df = pd.read_csv('Indias_Electricity_Consumption_Dataset.csv')

# 2. Convert Dates to actual datetime objects
# This handles the '06-01-2013' format
df['Dates'] = pd.to_datetime(df['Dates'], dayfirst=True)

# 3. Filter for the project requirements (Jan 2019 to Dec 5, 2020)
mask = (df['Dates'] >= '2019-01-01') & (df['Dates'] <= '2020-12-05')
df_filtered = df.loc[mask].copy()

# This turns 30+ columns into 3 columns: Date, State, and Consumption
# We exclude 'Sr' and 'Total Consumption' from the states list
state_columns = [col for col in df_filtered.columns if col not in ['Sr', 'Dates', 'Total Consumption']]

df_cleaned = pd.melt(df_filtered, 
                     id_vars=['Dates'], 
                     value_vars=state_columns,
                     var_name='State', 
                     value_name='Consumption_MU')

# Create a dictionary for professional names
name_map = {
    'Pondy': 'Puducherry',
    'UP': 'Uttar Pradesh',
    'HP': 'Himachal Pradesh',
    'MP': 'Madhya Pradesh',
    'J&K': 'Jammu & Kashmir',
    'DNH': 'Dadra and Nagar Haveli and Daman and Diu',
    'DD': 'Dadra and Nagar Haveli and Daman and Diu',
    'West Bengal': 'West Bengal'
}

# Apply the mapping to the State column
df_cleaned['State'] = df_cleaned['State'].replace(name_map)

# 5. Save the clean file
df_cleaned.to_csv('power_consumption_clean.csv', index=False)
print("Done! File 'power_consumption_clean.csv' is ready.")
```
